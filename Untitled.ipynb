# Titanic Data Analysis
Analyze survival patterns using Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")

df.head()
```

## Explore Dataset

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

## Check Missing Values

```python
df.isnull().sum()
```

## Data Cleaning
Handle missing values (especially Age).

```python
# Fill missing Age with median
df["Age"].fillna(df["Age"].median(), inplace=True)

# Drop Cabin (too many missing values)
df.drop("Cabin", axis=1, inplace=True)

# Fill Embarked with most frequent value
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

## Survival by Gender

```python
df.groupby("Sex")["Survived"].mean()
```

## Survival by Age Group

```python
# Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ["Child", "Teen", "Adult", "Mid-Age", "Senior"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

df.groupby("AgeGroup")["Survived"].mean()
```

## Visualization: Survival by Gender

```python
sns.barplot(x="Sex", y="Survived", data=df)

plt.title("Survival Rate by Gender")
plt.show()
```

## Visualization: Survival by Class

```python
sns.barplot(x="Pclass", y="Survived", data=df)

plt.title("Survival Rate by Passenger Class")
plt.show()
```

## Visualization: Age Distribution

```python
df["Age"].plot(kind="hist", bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

## Conclusion

- Females had higher survival rate than males  
- Passengers in 1st class had higher survival chances  
- Adults had moderate survival, children higher survival
