## Step 1: Import Required Libraries
In this step, we import the libraries required for data analysis and visualization.
- pandas → data manipulation
- matplotlib → data visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better visualization
sns.set_style("whitegrid")
```

## Step 2: Load the Dataset
We load the student performance dataset using pandas.

```python
# Load dataset
df = pd.read_csv("student-mat1.csv")

# Display first 5 rows
df.head()
```

## Step 3: Explore the Dataset
We inspect the dataset structure including:
- shape of dataset

- sample records

## Step 4: Check Missing Values
Missing values can affect analysis, so we check for them.

```python
df.isnull().sum()
```

## Step 5: Remove Duplicate Data
Duplicate data can create bias in analysis.

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()

df.shape
```

## Analysis 1: Average Final Grade
We calculate the average of final grades (G3).

```python
average_grade = df["G3"].mean()
average_grade
```

## Analysis 2: Students scoring above 15
We count how many students scored above 15 in the final grade.

```python
high_score = df[df["G3"] > 15]

len(high_score)
```

## Analysis 3: Study Time vs Performance
We check correlation between study time and final grade.

```python
correlation = df["studytime"].corr(df["G3"])
correlation
```

## Analysis 4: Gender Performance
We compare average final grades between male and female students.

```python
df.groupby("sex")["G3"].mean()
```

## Visualization 1: Histogram of Final Grades
Shows distribution of student grades.

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")

plt.show()
```

## Visualization 2: Study Time vs Grades
Scatter plot showing relationship between study time and final grade.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

## Visualization 3: Gender vs Average Score
Bar chart comparing average scores between male and female students.

```python
gender_avg = df.groupby("sex")["G3"].mean()

gender_avg.plot(kind="bar")

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```
